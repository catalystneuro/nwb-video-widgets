# NWB Video Widgets - Local NWB File Example

This notebook demonstrates streaming NWB video data directly from a local NWB file.

```python
from pathlib import Path

from pynwb import NWBHDF5IO, read_nwb

from tests.fixtures.synthetic_nwb import create_nwbfile_with_external_videos
from tests.fixtures.synthetic_video import create_synthetic_video

nwbfile_path = Path("/Users/pauladkisson/Documents/CatalystNeuro/DANDI/nwb-video-widgets/synthetic_nwbfile.nwb")
video_file_path = Path("/Users/pauladkisson/Documents/CatalystNeuro/DANDI/nwb-video-widgets/synthetic_video.mp4")

create_synthetic_video(video_file_path, num_frames=30, width=160, height=120)
nwbfile = create_nwbfile_with_external_videos({"VideoCamera": video_file_path})
with NWBHDF5IO(nwbfile_path, "w") as io:
    io.write(nwbfile)
nwbfile = read_nwb(nwbfile_path)
```

## Video Player

```python
from nwb_video_widgets import NWBLocalVideoPlayer

widget = NWBLocalVideoPlayer(nwbfile=nwbfile)
widget
```
