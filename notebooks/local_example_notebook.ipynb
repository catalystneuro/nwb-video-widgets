# NWB Video Widgets - Local NWB File Example

This notebook demonstrates streaming NWB video data directly from a local NWB file.

```python
from pathlib import Path

from pynwb import NWBHDF5IO, read_nwb

from nwb_video_widgets.testing.synthetic_nwb import create_nwbfile_with_external_videos

nwb_video_widgets_path = Path.cwd().parent
nwbfile_path = nwb_video_widgets_path / "tests" / "fixtures" / "videos" / "synthetic_nwbfile.nwb"
video_file_path = nwb_video_widgets_path / "tests" / "fixtures" / "videos" / "stub_h264.mp4"
if nwbfile_path.exists():
    nwbfile_path.unlink()

nwbfile = create_nwbfile_with_external_videos({"VideoCamera": video_file_path})
with NWBHDF5IO(nwbfile_path, "w") as io:
    io.write(nwbfile)
nwbfile = read_nwb(nwbfile_path)
```

## Video Player

```python
from nwb_video_widgets import NWBLocalVideoPlayer

widget = NWBLocalVideoPlayer(nwbfile=nwbfile)
widget
```

## Pose Estimation Widget

Stream video with pose estimation overlay.

```python
from pathlib import Path

from pynwb import NWBHDF5IO, read_nwb

from nwb_video_widgets.testing.synthetic_nwb import create_nwbfile_with_videos_and_pose

nwb_video_widgets_path = Path.cwd().parent
nwbfile_path = nwb_video_widgets_path / "tests" / "fixtures" / "videos" / "synthetic_nwbfile.nwb"
video_file_path = nwb_video_widgets_path / "tests" / "fixtures" / "videos" / "stub_h264.mp4"
if nwbfile_path.exists():
    nwbfile_path.unlink()

nwbfile = create_nwbfile_with_videos_and_pose(
    video_paths={"VideoCamera": video_file_path},
    camera_names=["VideoCamera"],
    keypoint_names=["Keypoint1", "Keypoint2"],
    num_frames=30,
)
with NWBHDF5IO(nwbfile_path, "w") as io:
    io.write(nwbfile)
nwbfile = read_nwb(nwbfile_path)
```

```python
from nwb_video_widgets import NWBLocalPoseEstimationWidget

widget = NWBLocalPoseEstimationWidget(nwbfile=nwbfile)
widget
```
